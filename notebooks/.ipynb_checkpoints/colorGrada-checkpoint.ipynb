```python
fn = "C:\\Users\\nsdev\\code\\web-dev\\aectech25\\backend\\CalculatedValues.csv"

with open(fn, 'r') as f:
    data = f.readlines()
id_arr=[]
num_arr=[]
sum = 0
for i, line in enumerate(data): 
    if i>10:
        break
    arr = line.split(',')
    # print(arr)
    _id = arr[4]
    num = arr[7].split("\n")[0]
    num_arr.append(num)
    id_arr.append(_id)
    sum += num

max_num = max(num_arr)
min_num = min(num_arr)
max_num
print(f'max_num  = {max_num}, min_num = {min_num}')

```
